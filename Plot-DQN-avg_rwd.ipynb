```python
import matplotlib.pyplot as plt


lr_15 = []
episodes_15 = []
count = 0
lr_1 = []
episodes_1 = []
lr_2 = []
episodes_2 = []

with open('C:/Users/yonat/OneDrive/Desktop/Reinforcment Learning/Project/DQN-0.0015,64,64.txt') as f:
    for line in f:
        line = line.split('=')
        line[2] = line[2].split(',')
        #print(float(line[2][0].strip()))
        lr_15.append(float(line[3].strip()))
        episodes_15.append(count)
        count += 1
print(count)
count=0
with open('C:/Users/yonat/OneDrive/Desktop/Reinforcment Learning/Project/DQN-0.001-64-6.txt') as f:
    for line in f:
        line = line.split('=')
       # line[2] = line[2].split(',')
       # print(float(line[2][0].strip()))
        lr_1.append(float(line[3].strip()))
        episodes_1.append(count)
        count += 1
print(count)
count=0
with open('C:/Users/yonat/OneDrive/Desktop/Reinforcment Learning/Project/DQN-0.002-64-64.txt') as f:
    for line in f:
        line = line.split('=')
       # line[2] = line[2].split(',')
        #print(float(line[2][0].strip()))
        lr_2.append(float(line[3].strip()))
        episodes_2.append(count)
        count += 1
print(count)

fig = plt.figure(figsize=[15,10])
ax = fig.add_subplot(111)
ax.set_title("Acquired Rewards")
plt.plot(episodes_15, lr_15, 'r', label='DQN with lr = 0.0015')
plt.plot(episodes_1, lr_1, 'g', label='DQN with lr = 0.001')
plt.plot(episodes_2, lr_2, 'b', label='DQN with lr = 0.002')
plt.grid()
ax.set_xlabel("episodes")
ax.set_ylabel("Average Reward")
ax.legend(loc='best')
plt.savefig('dqn_lr_comparison_avg_reward.jpg')
plt.show()
```
